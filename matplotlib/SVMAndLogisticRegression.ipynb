根据机器学习吴恩达老师学习笔记,SVM与logistic regression cost function关系

![](../photo/svmlogistic.jpg)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


# 数据
z = np.linspace(-3,3,1000)
y1 = -(np.log(1/(1+np.exp(-z))))
y2 = -np.log(1-(1/(1+np.exp(-z))))
cost1_y = np.array([])
cost2_y = np.array([])
def cost1(i):
    if i <= 1:
        return -2*i + 2
    else:
        return 0
def cost2(i):
    if i <= -1:
        return 0
    else:
        return 2*i+2
for i in z:
    cost1_y = np.append(cost1_y,cost1(i))
    cost2_y = np.append(cost2_y,cost2(i))
plt.figure(figsize=(10,10))

# print(cost1_y)
plt.subplot(211)
plt.title('logistic regression cost function')
plt.xlim(-3,3)
plt.ylim(0,3)
plt.xlabel('z')
plt.ylabel('y')
plt.plot(z,y1,label="y=1")
plt.plot(z,y2,label='y=0',color='red')
plt.legend()

# SVM y = 1
plt.subplot(223)
plt.title('y=1,z>=1')
# plt.xlim(-3,3)
# plt.ylim(0,3)
plt.xlabel('z')
plt.ylabel('y')
plt.plot(z,cost1_y,label="y=1")
plt.scatter(1,0,color='red')
plt.annotate(r'$(1,0)$', xy=(1, 0), xycoords='data', xytext=(-30, +30),
             textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle='->', connectionstyle="arc3,rad=.2"))

# SVM y = 0
plt.subplot(224)
plt.title('y=0,z<=-1')
# plt.xlim(-3,3)
# plt.ylim(0,3)
plt.xlabel('z')
plt.ylabel('y')
plt.plot(z,cost2_y,label="y=1")
plt.scatter(-1,0,color='red')
plt.annotate(r'$(-1,0)$', xy=(-1, 0), xycoords='data', xytext=(-30, +30),
             textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle='->', connectionstyle="arc3,rad=.2"))

plt.tight_layout()
plt.show()
```
